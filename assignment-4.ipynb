# Bank Dataset

```python
ds=pd.read_csv(r"C:\Users\sanju\Downloads\Telegram Desktop\bank.csv")
```

```python
ds
```

```python
ds.isnull().any()
```

```python
ds["marital"].unique()
```

```python
def t_marital(a):
    if a=="married":
        return 0
    if a=="single":
        return 1
    if a=="divorced":
        return 2
```

```python
ds["marital"]=ds["marital"].apply(t_marital)
ds
```

```python
ds["job"].unique()
```

```python
def t_job(a):
    if a=="admin.":
        return 0
    if a=="technician":
        return 1
    if a=="services":
        return 2
    if a=="management":
        return 3
    if a=="retired":
        return 4
    if a=="blue-collar":
        return 5
    if a=="unemployed":
        return 6
    if a=="entrepreneur":
        return 7
    if a=="housemaid":
        return 8
    if a=="unknown":
        return 9
    if a=="self-employed":
        return 10
    if a=="student":
        return 11
```

```python
ds["job"]=ds["job"].apply(t_job)
ds
```

```python
ds["education"].unique()
```

```python
def t_edu(a):
    if a=="secondary":
        return 0
    if a=="tertiary":
        return 1
    if a=="primary":
        return 2
    if a=="unknown":
        return 3
```

```python
ds["education"]=ds["education"].apply(t_edu)
ds
```
